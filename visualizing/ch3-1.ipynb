### Simple Plot Line Chart

```python
years = [x for x in range(1950,2020,10)]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
years
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.show()
```

### Simple Bar Chart

```python
movies = ["Aniie Hall", "Ben-Hur", "Casbalnca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.bar(range(len(movies)), num_oscars)
plt.title("My favourite Movies")
plt.ylabel("Number of Academy Awards")
plt.xticks(range(len(movies)), movies)
plt.show()
```

#### Using Bar Chart for Histogram

```python
from collections import Counter
import random
grades = random.sample(range(10, 101), 15)
grades
```

```python
histogram = Counter(min((grade//10)*10, 90) for grade in grades)
histogram
```

```python
plt.bar([x+5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor=(0,0,0))
plt.axis([-5, 105, 0, 5])

plt.xticks([10*i for i in range(0,11)])
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of Exam I grades")
plt.show()
```

### Scatter Plots

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
import string
labels = list(string.ascii_lowercase[:9])
labels
```

```python
plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy = (friend_count, minute_count),
                 xytext = (5, -5),
                textcoords='offset points')
plt.title("Daily Minutes vs No of Friends")
plt.xlabel("No of Friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

```python
plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(f'{label}({friend_count},{minute_count})', #label was a string, you can put a different string
                 xy = (friend_count, minute_count),
                 xytext = (5, -5),
                textcoords='offset points')
plt.title("Daily Minutes vs No of Friends")
plt.xlabel("No of Friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```
